#### Caracteristicas Vistas (Views) DDL

![views.jpg](./views.jpg "views.jpg")

```python
%sql
-- Creamos una vista que transforma datos "al vuelo"
CREATE OR REPLACE VIEW ecomotive.bronze.v_drivers_cleaned AS
SELECT 
    driver_id,
    raw_name AS driver_name
FROM ecomotive.bronze.bronze_drivers;

-- Comprobamos el resultado (BI verá esto)
SELECT * FROM ecomotive.bronze.v_drivers_cleaned;
```

```python
%sql
-- Creamos la vista temporal (nota: sin esquema 'ecomotive.bronze')
CREATE OR REPLACE TEMP VIEW temp_hazmat_drivers AS
SELECT 
    driver_id,
    raw_name,
    certifications_list
FROM ecomotive.bronze.bronze_drivers;

-- Consultamos dentro de esta sesión
SELECT * FROM temp_hazmat_drivers;
```

```python
%sql
-- Creamos la vista global
CREATE OR REPLACE GLOBAL TEMP VIEW gv_new_hires_2023 AS
SELECT 
    driver_id,
    raw_name,
    hired_date
FROM ecomotive.bronze.bronze_drivers
WHERE hired_date >= '2023-01-01';

-- IMPORTANTE: Para leerla hay que usar el prefijo 'global_temp'
SELECT * FROM global_temp.gv_new_hires_2023;
```

```python
%sql
USE CATALOG ecomotive;
SHOW VIEWS IN ecomotive.bronze;
```

- Standard: v_drivers_cleaned
- Temproal: temp_hazmat_drivers
- Global Temporal: gv_new_hires_2023
