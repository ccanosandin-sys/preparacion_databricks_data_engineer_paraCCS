```python
from pyspark.sql.functions import col

# 1. Creamos dos DataFrames medianos
# 5000 filas en cada lado no parece mucho...
df1 = spark.range(5000).withColumnRenamed("id", "id1")
df2 = spark.range(5000).withColumnRenamed("id", "id2")

# 2. La Operación Nuclear: CROSS JOIN
# 5,000 x 5,000 = 25,000,000 (25 Millones de filas resultantes)
# Esto obligará a Spark a trabajar duro sin necesidad de trucar la configuración.
print("⏳ Iniciando operación pesada... (Puede tardar unos segundos/minutos)")
result = df1.crossJoin(df2)

# 3. Acción: Contar los resultados (fuerza el cálculo)
count = result.count()

print(f"✅ Terminado. Filas procesadas: {count}")
```
