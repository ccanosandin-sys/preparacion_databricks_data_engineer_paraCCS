```python
%sql
-- Ver quién tiene llaves de la tabla
SHOW GRANTS ON TABLE ecomotive.gold.dim_drivers;
```

```python
%sql
-- 1. Abrir la puerta del Edificio (Obligatorio para entrar)
GRANT USE CATALOG ON CATALOG ecomotive TO `account users`;

-- 2. Abrir la puerta del Pasillo (Obligatorio para llegar a la tabla)
GRANT USE SCHEMA ON SCHEMA ecomotive.gold TO `account users`;

-- 3. Finalmente, dar permiso de lectura (Solo mirar, no tocar)
GRANT SELECT ON TABLE ecomotive.gold.dim_drivers TO `account users`;
```

```python
%sql
-- Dar permiso de escritura/modificación
GRANT MODIFY ON TABLE ecomotive.gold.dim_drivers TO `account users`;
```

```python
%sql
SHOW GRANTS ON TABLE ecomotive.gold.dim_drivers;
```

```python
%sql
-- Quitar permisos específicos
REVOKE MODIFY ON TABLE ecomotive.gold.dim_drivers FROM `account users`;

-- O quitarlo todo (Opción Nuclear)
REVOKE ALL PRIVILEGES ON TABLE ecomotive.gold.dim_drivers FROM `account users`;
```
